Next, we need to re-scale the input **Golden Curve** to match and have the same amount of data points as the **Calibration Shot Data** *(via Cubic Spline interpolation)*

* Note, I converted the first row from **mm** to **cm**, will have to add functionality to determine if the information is already in **mm** or **cm**.

```python
golden_curve.iloc[0]
```

```python
interpol_step = (golden_curve.iloc[0].max() - golden_curve.iloc[0].min()) / 179
gc_interpolated_depth = np.arange(golden_curve.iloc[0].min(),golden_curve.iloc[0].max()+interpol_step,interpol_step)
```

```python
gc_cspline = CubicSpline(golden_curve.iloc[0],golden_curve.iloc[1])
gc_cspline(gc_interpolated_depth)
```

```python
gc_cspline = CubicSpline(golden_curve.iloc[0],golden_curve.iloc[1])
gc_uspline = UnivariateSpline(golden_curve.iloc[0],golden_curve.iloc[1],k=3,s=0.0002)
gc_cs_temp = gc_cspline(gc_interpolated_depth)
gc_us_temp = gc_uspline(gc_interpolated_depth)
plt.figure(figsize=(15,15), dpi=80)
plt.plot(golden_curve.iloc[0],golden_curve.iloc[1],'bo',label='Golden Curve Master')
plt.plot(gc_interpolated_depth, gc_cs_temp,'k-',label='Golden Curve Cubic Spline')
plt.plot(gc_interpolated_depth, gc_us_temp,'k--',label='Golden Curve Quartic UniV Spline')
plt.xlim(0,350)
plt.xlabel('Proton Depth [mm]')
plt.ylabel('Uniformity [%]')
plt.legend()
plt.show()
```

These results clearly show that the standard Cubic Spline seems to have a much better result than forcing a Univariate Spline with k=3 (so also a cubic spline) and a great deal of knots gives poor results after the distal end of the Bragg Peak.

```python
%%latex

Next, we calculate the Uniformity Gains in full. We do this with the following formulas:
\begin{equation*}
\\
MLIC~Normalization~(i) = \frac{MLIC~Raw~Data~(i)}{Max(MLIC~Raw~Data)}
\\
Uniformity~Gains~(i) = \frac{Golden~Curve~(i)}{MLIC~Normalization~(i)}
\\
\end{equation*}

**Note:** We need to also ensure that, due to interpolation, there are no **negative** signal values in the golden curve, so we will call the absolute value of the array onto itself.
```

```python
cal_normal = cal_shot_data.iloc[0] / cal_shot_data.iloc[0].max()
Uniformity_gains = (np.absolute(gc_cs_temp) / cal_normal) / 100
```

```python
plt.figure(figsize=(15,15), dpi=80)
plt.plot(list(range(0,180)),Uniformity_gains,'bo',label='Uniformity Gains')
plt.xlim(0,179)
plt.xlabel('Channel Number')
plt.ylabel('Uniformity Gains [%]')
plt.title('Uniformity Gains')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,15), dpi=80)
plt.plot(gc_interpolated_depth,cal_normal*100,'bo',label='MLIC Raw Data')
plt.plot(gc_interpolated_depth, gc_cs_temp,'ro-',label='Golden Curve Interpolated')
plt.xlim(0,350)
plt.xlabel('Proton Depth [mm]')
plt.ylabel('Signal [%]')
plt.legend()
plt.show()
```
